#  Step-by-Step Explanation of the BB84 Quantum Key Distribution Protocol

Let's walk through the **BB84 protocol** in a structured, detailed manner.

---

### Step 1: Alice Prepares the Qubits

- Alice begins by selecting two random bit strings of length $n$:
  - $k = (k_1, k_2, ..., k_n)$ â†’ The actual bits she wants to transmit (this will become part of the final key).
  - $b = (b_1, b_2, ..., b_n)$ â†’ Determines which **basis** she will use to encode each qubit.

---

### The Two Possible Bases:

1. **Computational (Rectilinear) Basis** ($\{ \ket{0}, \ket{1} \}$):
   - Used when $b_i = 0$
   - $k_i = 0 \Rightarrow \ket{0}$, $k_i = 1 \Rightarrow \ket{1}$

2. **Diagonal (Hadamard) Basis** ($\{ \ket{+}, \ket{-} \}$):
   - Used when $b_i = 1$
   - Where:

     $$
     \ket{+} = \frac{1}{\sqrt{2}} (\ket{0} + \ket{1}) \\
     \ket{-} = \frac{1}{\sqrt{2}} (\ket{0} - \ket{1})
     $$

   - $k_i = 0 \Rightarrow \ket{+}$, $k_i = 1 \Rightarrow \ket{-}$

### Visualizing the Encoding:

This table summarizes how each bit $k_i$ is encoded depending on the basis $b_i$:

|               | $b_i = 0$ (Rectilinear)     | $b_i = 1$ (Diagonal)     |
|---------------|-----------------------------|---------------------------|
| $k_i = 0$     | $\ket{0}$                   | $\ket{+}$                |
| $k_i = 1$     | $\ket{1}$                   | $\ket{-}$                |

**Interpretation:** Alice now has a sequence of $n$ encoded qubits:  
$q_1, q_2, ..., q_n$ â€” each being one of $\ket{0}, \ket{1}, \ket{+}, \ket{-}$

---

## Step 2: Alice Sends the Qubits to Bob

- Alice transmits all $n$ encoded qubits to Bob via a quantum channel.

- Simultaneously, Bob generates his **own random basis string**  
  $\tilde{b} = \tilde{b}_1, \tilde{b}_2, ..., \tilde{b}_n$, where:
  - $\tilde{b}_i = 0$ means Bob measures qubit $q_i$ in the rectilinear basis.
  - $\tilde{b}_i = 1$ means Bob measures qubit $q_i$ in the diagonal basis.

- Bob records the **outcomes** of his measurements in a string  
  $\tilde{k} = \tilde{k}_1, \tilde{k}_2, ..., \tilde{k}_n$


## Step 3: Basis Comparison and Sifting

- After transmission, Alice and Bob **publicly share** their basis strings $b$ and $\tilde{b}$ (but *not* their bit values).

- For each position $i$, they compare $b_i$ and $\tilde{b}_i$:

  - If $b_i = \tilde{b}_i$:  
    Bob measured in the **same basis** that Alice used to encode â†’ measurement outcome is **correct with high probability**.  
    So, retain $\tilde{k}_i$.

  - If $b_i \neq \tilde{b}_i$:  
    Bob used a **different basis**, and the outcome is essentially **random** â†’ discard $\tilde{k}_i$.

- After discarding mismatched basis positions, Alice and Bob are left with a shorter, **synchronized bit string** derived from:

$$
\text{Final Key: } k_i = \tilde{k}_i \quad \text{for all } i \text{ where } b_i = \tilde{b}_i
$$

###  Step 4: Error Detection & Correction

Even in ideal conditions, **noise** or potential **eavesdropping** can cause discrepancies between Alice and Bobâ€™s sifted keys. To ensure they end up with identical keys, they perform two steps:


####  Error Detection (QBER Estimation)

- Alice and Bob randomly select a small subset of the sifted key and **publicly compare** those bits.
- This allows them to compute the **Quantum Bit Error Rate (QBER)**:

$$
\text{QBER} = \frac{\text{Number of mismatches}}{\text{Number of test bits}}
$$

- If QBER exceeds a predefined threshold (commonly **11%** in BB84), it indicates a possible eavesdropping attempt, and the protocol is **aborted**.


#### ðŸ”§ Error Correction

- If QBER is acceptable, Alice and Bob perform **error correction** to reconcile their remaining sifted key.

- This is done via **classical error correction protocols** over a **public authenticated channel**. Common techniques include:
  - **Cascade protocol** (interactive, iterative parity-check-based)
  - **LDPC codes** (Low-Density Parity-Check)
  - **Winnow**, **Hamming**, or **Turbo codes** in some implementations


####  Simulation Note

In our BB84 simulation:

- We assume **perfect error correction** with a correction **efficiency factor** (e.g., 0.8), which models how many bits are lost during correction.
- Only a fraction of the sifted key survives after this step.

Example:
If sifted key length = 1000 bits and correction efficiency = 0.8,  
then final key after correction â‰ˆ **800 bits**


####  Important

Error correction must reveal **only minimal information** to Eve. Thatâ€™s why this step is followed by **privacy amplification**, which removes any residual leak.


 **Summary**:  
Error correction ensures that Alice and Bob end up with **identical keys**, despite noise or tampering. The trade-off is that a portion of the key is consumed to guarantee this alignment.

###  Step 5: Privacy Amplification

To eliminate any partial information that an eavesdropper (Eve) might have gained during transmission, Alice and Bob perform **privacy amplification** â€” a process that compresses their shared key into a shorter, **secret** key that is provably secure.


####  Why Privacy Amplification?

Even after error correction, Eve might still possess partial knowledge about the raw key due to:

- Detector noise  
- Multi-photon pulses (vulnerable to photon-number-splitting attacks)  
- Side-channel leaks  
- Active interception attempts

Privacy amplification ensures that **any such knowledge becomes useless**, by reducing the key length in a mathematically safe way.


####  Methods of Privacy Amplification

- In our **simulation**, we used a **simple XOR-based method**, where adjacent bits are combined:
  
  Example:  
  Original bits: `[1, 0, 1, 1, 0, 0, 1, 1]`  
  Amplified: `[1âŠ•0, 1âŠ•1, 0âŠ•0, 1âŠ•1] = [1, 0, 0, 0]`

  This is a basic and illustrative method that mimics how entropy is compressed.

- In **real QKD systems**, more sophisticated techniques are used, such as:
  - **Universal hashing functions** (e.g., Toeplitz matrices)
  - **Random extractors** from complexity theory
  - **Hash-based protocols** proven to be secure under the **Leftover Hash Lemma**

These methods provide **information-theoretic security**, meaning the final key remains secure **even against adversaries with unlimited computational power**.


####  How Final Key Length Is Determined

- The **length of the final key** after privacy amplification depends on:
  - The **estimated knowledge Eve might have** (from QBER and other parameters)
  - The **error correction efficiency**
  - The **confidence level** in the security proof (usually chosen by the protocol)

- In general:
  - **Lower QBER â†’ longer final key**
  - **Higher noise or detected eavesdropping â†’ shorter final key**
  - The protocol always ensures that **the leftover key is safe**, regardless of compression.


 **Bottom line**: Privacy amplification shrinks the key just enough to erase any trace of Eveâ€™s potential knowledge â€” leaving Alice and Bob with a smaller, but truly secure, shared secret.

## Eavesdropping Detection

If an **eavesdropper (Eve)** intercepts and resends the qubits:

- She **does not know** Aliceâ€™s basis choices.
- Measuring the qubit will collapse its state.
- When Eve sends a new qubit to Bob, it may not match Aliceâ€™s original encoding.
- This increases the **Quantum Bit Error Rate (QBER)**

A higher-than-expected QBER reveals **eavesdropping** activity, prompting Alice and Bob to abort the protocol or run **error correction and privacy amplification** steps.

## Summary of BB84 Protocol

1. Alice randomly selects bit string $k$ and basis string $b$
2. Encodes each bit $k_i$ into photon $q_i$ using basis $b_i$
3. Sends the photons $q_i$ to Bob
4. Bob measures each $q_i$ using his random basis $\tilde{b}_i$
5. Alice and Bob reveal their basis strings $b$, $\tilde{b}$
6. Keep only positions where $b_i = \tilde{b}_i$
7. Form the final shared key using the corresponding $k_i$ and $\tilde{k}_i$  
   â†’ Ensuring **unconditional security**

8. Estimate the **Quantum Bit Error Rate (QBER)** by comparing a subset of the sifted key  
   â†’ If QBER is too high (typically >11%), the protocol is aborted

9. Apply **error correction protocols** (like Cascade or LDPC) to reconcile discrepancies  
   â†’ Some bits are discarded during this step to ensure both parties hold identical keys

10. Perform **privacy amplification** using methods like hashing or XORing adjacent bits  
    â†’ Compress the reconciled key to remove any information potentially leaked to an eavesdropper

The final result is a shorter, but **provably secure** key shared between Alice and Bob.
